### (1) np.where 활용

```python
import pandas as pd
import numpy as np
 
# CSV 파일을 읽어 DataFrame 변수에 저장하기
customerData = pd.read_csv("/content/drive/MyDrive/0430/customerdata.csv")
display(customerData.head(10))

customerData.PRODUCTAGE=np.where(customerData.PRODUCTAGE <= 1, 
                                 # productage 1보다 작거나 같으면면 1로 변경, 
                                 # 아니면 np.where()한번 더 사용 
                                 # => 2보다 작거나 같으면 2로 변경, 아니면 3으로
                                 1,np.where(customerData.PRODUCTAGE <= 2, 2, 3))

customerData.head()
```

### (2) loc 활용

```python
# %%timeit
import pandas as pd
 
# CSV 파일을 읽어 DataFrame 변수에 저장하기
customerData = pd.read_csv("/content/drive/MyDrive/0430/customerdata.csv")

# 각 조건에 맞게 NEW_AGE 컬럼 값을 1 or 2 or 3으로 변경
condition1 = customerData.PRODUCTAGE <= 1
condition2 = (customerData.PRODUCTAGE > 1) & \
             (customerData.PRODUCTAGE <= 2)
condition3 = customerData.PRODUCTAGE > 2
customerData.loc[ condition1, "NEW_AGE" ] = 1
customerData.loc[ condition2, "NEW_AGE" ] = 2
customerData.loc[ condition3, "NEW_AGE" ] = 3
customerData.head(10)
```

# 7. 데이터 그룹별 집계함수

```python
import pandas as pd
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_csv("../dataset/customerdata.csv")
customerData.head()
```

무이자할부(EMI) 기준 평균지출비용(AVGPRICE)의 평균 산출

```python
# 무이자할부(EMI) 기준 평균지출비용(AVGPRICE)의 평균 산출
groupKey = ["EMI"]
meanValue = customerData.groupby(by=groupKey)["AVGPRICE"].agg(["mean"])
print(meanValue)
# 컬럼명 변경 (mean -> MEAN_AVGPRICE)
meanValue.rename(columns={"mean":"MEAN_AVGPRICE"}, inplace=True)
# 인덱스 초기화
meanValue.reset_index(inplace=True)
meanValue
```

```python
# 무이자할부(EMI) 기준 평균지출비용(AVGPRICE)의 사이즈 산출
groupKey = ["EMI"]
meanValue = customerData.\
groupby(by=groupKey)\
        ["AVGPRICE"].\
        agg(["count"])

# 컬럼명 변경 (size -> SIZE_CS)
meanValue.rename(columns={"size":"SIZE_CS"}, 
                 inplace=True)
# 인덱스 초기화
meanValue.reset_index(inplace=True)
meanValue
```

# 8. 고급 연산

```python
import pandas as pd

# 데이터 불러오기
kopoData = pd.read_csv("../dataset/kopo_product_volume.csv")

# 데이터 선택 (ST0001 제품)
kopoDataA01 = kopoData.loc[ kopoData.PRODUCTGROUP == "ST0001" ]

# 데이터 정렬
sortKey = ["REGIONID","PRODUCTGROUP","YEARWEEK"]
kopoSortedData = kopoDataA01.sort_values(sortKey, ignore_index=True)
kopoSortedData.head()
```

```python
# 이동평균 산출 (구간:5, 중심기준)
kopoSortedData["MA5"] = kopoSortedData["VOLUME"].rolling(window=5, 
                                 center=True, 
                                 min_periods=1).mean()
kopoSortedData.head()
```

```python
kopoSortedData["EFFECT"] = kopoSortedData.VOLUME / \
kopoSortedData.MA5
kopoSortedData.head()
```

```python
# 주차컬럼 생성
kopoSortedData["WEEK"] = kopoSortedData["YEARWEEK"].astype(str).str[4:6]

# 지역, 상품, 주차별 효과 산출
groupKey = ["REGIONID","PRODUCTGROUP","WEEK"]
weeklyEffect = kopoSortedData.\
                groupby(by=groupKey)["EFFECT"].\
                agg(["mean"]).reset_index()
weeklyEffect.rename(columns={"mean":"WEEKLY_EFFECT"}, inplace=True)
weeklyEffect.head(2)
```
